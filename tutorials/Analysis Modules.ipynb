# Introduction
This tutorial gives an overview of the modules in ``PhoPro.analysis``:

* ``artifact`` - handles the detection and correction of spike and jump artifacts.

* ``comparison`` - provides cluster-based permutation tests for finding the points where groups of timeseries statistically differ.

* ``peaks`` - contains peak detection methods that automatically calculate peak metrics.

* ``FMM`` - provides an API for Functional Linear Mixed Modeling as implemented in the ``fast-fmm-rpy2`` package, from [Loewinger et al, 2025](https://doi.org/10.7554/eLife.95802.2).
